# diabets

## import Dataset



```
from sklearn import datasets
```





````python
diabets = datasets.load_diabetes()
````

````python
diabets
````

````python
X = diabets.data
Y = diabets.target
````

````python
X.shape , Y.shape
````

# **Train and split**

````python
from sklearn.model_selection import train_test_split
````

````python
X_train , X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
````

# Evaluate Model

````python
from sklearn.metrics import mean_squared_error , r2_score
````

````python
model = LinearRegression()
````

````python
model.fit(X_train,Y_train)
````

````python
y_predict = model.predict(X_test)
````

````python
print(f"coefficient:  {model.coef_}")
print(f"Model intercept:  {model.intercept_}")
print(f"MSE: {mean_squared_error(Y_test,y_predict)}")
print(f"R2: {r2_score(Y_test,y_predict)}")
````

The R2 is kinda low for these Prdiction ,

- if the R2 IS closer to 1 mean model is perfectly prediction
- if the R2 is far away from 1 mean model is bad performance t predict

# **Visulization**

````python
plt.scatter(Y_test,y_predict, color ='k')
plt.plot(Y_test, Y_test, color='r')
````

Import Libary

````python
import pandas as pd
````

````python
BostonHousing = pd.read_csv("BostonHousing.csv")
BostonHousing
````

````python
Y = BostonHousing.medv
X = BostonHousing.drop(["medv"],axis= 1)
````

## import Linear Regression model and Metric


````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
````

##Train and Split


```
X_train , X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```



````python
X_train , X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
````

## Evaluate model

````python
model = LinearRegression()
````

## Fit the model

````python
model.fit(X_train,Y_train)
````

````python
Y_pre = model.predict(X_test)
````

````python
print(f"coefficient:  {model.coef_}")
print(f"Model intercept:  {model.intercept_}")
print(f"MSE: {mean_squared_error(Y_test,Y_pre)}")
print(f"R2: {r2_score(Y_test,Y_pre)}")
print(model.score(X_test, Y_test))
````
